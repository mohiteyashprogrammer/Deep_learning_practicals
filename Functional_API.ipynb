```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import *
```

```python
x = Input(shape = (3,))

hidden1 = Dense(128,activation="relu")(x) # hideen layer1

hidden2 = Dense(64,activation="relu")(hidden1)

output1 = Dense(1,activation="linear")(hidden2)

output2 = Dense(1,activation="sigmoid")(hidden2)
```

```python
model = Model(inputs=x,outputs = [output1,output2])
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model,show_shapes=True)
```

```python
inputA= Input(shape = (32,))
inputB = Input(shape = (128,))

# i will create branch for inputA
x = Dense(8,activation="relu",kernel_initializer="he_normal")(inputA)

x1 = Dense(4,activation="relu",kernel_initializer="he_normal")(x)
```

```python
y = Dense(64,activation="relu",kernel_initializer="he_normal")(inputB)

y1 = Dense(32,activation="relu",kernel_initializer="he_normal")(y)

y2 = Dense(4,activation="relu",kernel_initializer="he_normal")(y1)
```

```python
# concanate this 2 branches
combine_value = concatenate([x1,y2])

z = Dense(2,activation="relu",kernel_initializer="he_normal")(combine_value)

z1 = Dense(1,activation="linear")(z)
```

```python
model = Model(inputs=[inputA,inputB],outputs=z1)
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model,show_shapes=True)
```

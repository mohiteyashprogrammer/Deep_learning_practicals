```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Mahender1998/Bank-customer-churn-prediction/main/churn.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

```python
## Drop The  Unwanted Columns
data.drop(["RowNumber","CustomerId","Surname"],axis=1,inplace=True)
```

```python
data.head()
```

```python
data["Geography"].value_counts()
```

```python
data["Gender"].value_counts()
```

```python
## Apply One Hot Encoding
data = pd.get_dummies(data,columns=["Geography","Gender"],drop_first=True)
```

```python
data.head()
```

```python
# Saprate Dependent and Indipendent variable
x = data.drop("Exited",axis=1)
y = data["Exited"]
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential   ## Using  sequential approach
from tensorflow.keras.layers import Dense  ## Help To Creat Hidden Layer
from keras.api._v2.keras import activations
```

```python
model = Sequential()
model.add(Dense(11,activation="sigmoid",input_dim=11)) ## Adding Layer  INput Layer
model.add(Dense(11,activation="sigmoid"))## Hidden Layer
model.add(Dense(1,activation="sigmoid"))## Output Layer
```

```python
model.summary()
```

```python
## Split data in to train test split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler  ## Scale Data Using Grident Disent
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## Compile Model 
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train,y_train,batch_size=65,epochs=100,verbose=1)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict = y_predict.argmax(axis = -1)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
```

```python
print(accuracy_score(y_test,y_predict))
print(roc_auc_score(y_test,y_predict))
print(classification_report(y_test,y_predict))
```
